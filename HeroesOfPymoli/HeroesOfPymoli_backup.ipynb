## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Calculate number of unique items
Number_of_Unique_Items = purchase_data_df["Item ID"].nunique()

# Calculate the average price 
Average_Price_Items = purchase_data_df["Price"].mean()

# Calculate the Number of Purchases 
Number_of_Purchases = purchase_data_df["Price"].count()

# Calculate the Total Revenue 
Total_Revenue = purchase_data_df["Price"].sum()

# Create Summary DataFrame
Summary_Purchasing_Analysis_df = pd.DataFrame({"Number of Unique Items": [Number_of_Unique_Items],
                              "Average Price": Average_Price_Items,
                              "Number of Purchases": Number_of_Purchases,
                              "Total Revenue": Total_Revenue})
# Format the values and left align
Summary_Purchasing_Analysis_df["Average Price"] = Summary_Purchasing_Analysis_df["Average Price"].astype(float).map("${:,.2f}".format)
Summary_Purchasing_Analysis_df["Total Revenue"] = Summary_Purchasing_Analysis_df["Total Revenue"].astype(float).map("${:,.2f}".format)
Summary_Purchasing_Analysis_df.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# purchase_data_df.head()

# Create df with just gender information
player_gender_df = purchase_data_df[["Gender","SN"]]
# player_gender_df.head()

# Create a dataframe grouped by gender.
player_gender_group = player_gender_df.groupby(["Gender"])
player_gender_group.head()

# Create summation of purchases by gender
player_gender_comparison_df = player_gender_group.nunique()
player_gender_comparison_df.head()

# Add the total of the column to each entry
player_gender_comparison_df["Total"] = player_gender_comparison_df["SN"].sum(axis=0)
player_gender_comparison_df.head()


# Clean up the Table column names
#Rename the summary statistic
player_gender_comparison_df=player_gender_comparison_df.rename(columns = {'SN':'Total Count'})

# Renmove the index name as shown in the sample (just to be exact!!)
# player_gender_comparison_df.index.name = None

# Add the Percentage and format it
player_gender_comparison_df["Percentage of Players"] = ((player_gender_comparison_df["Total Count"] / player_gender_comparison_df["Total"])*100)
player_gender_comparison_df["Percentage of Players"] = player_gender_comparison_df["Percentage of Players"].astype(float).map("{:,.2f}%".format)
player_gender_comparison_df.head()

# Sort Descending Total Count of purchases
sorted_gender_comparison_df = player_gender_comparison_df.sort_values(["Total Count"], ascending=False)

# Drop the Total column
del sorted_gender_comparison_df['Total']

sorted_gender_comparison_df.head()
```

## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Create df with just gender information
Purchase_gender_df = purchase_data_df[["Gender","Price"]]
# Purchase_gender_df.head()

# Create a dataframe grouped by gender.
Purchase_gender_group = Purchase_gender_df.groupby(["Gender"])
# Purchase_gender_group.head()

# count the purchases by gender 
Purchase_gender_stats = Purchase_gender_group.count()
# Purchase_gender_stats.head()

#Rename the summary statistic
Purchase_gender_stats=Purchase_gender_stats.rename(columns = {'Price':'Purchase Count'})
# Purchase_gender_stats.head()

# add the sum of the purchases by gender 
Purchase_gender_stats["Total Purchase Value"] = Purchase_gender_group.sum()
# Purchase_gender_stats.head()


Purchase_gender_stats["Average Purchase Price"] =  (Purchase_gender_stats["Total Purchase Value"] /Purchase_gender_stats["Purchase Count"])
# Purchase_gender_stats.head()

# Reorder the columns
Purchase_gender_stats = Purchase_gender_stats[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]
# Purchase_gender_stats.head()

# Merge two dataframes using an inner join
merge_player_info_df = pd.merge(Purchase_gender_stats, sorted_gender_comparison_df, on="Gender")

merge_player_info_df["Avg Total Purchase per Person"] = ((merge_player_info_df["Purchase Count"] /merge_player_info_df["Total Count"]) * merge_player_info_df["Average Purchase Price"])
# Format .map("${:,.2f}".format) 


# Add formatting
merge_player_info_df["Average Purchase Price"] = merge_player_info_df["Average Purchase Price"].map("${:,.2f}".format) 
merge_player_info_df["Total Purchase Value"]   = merge_player_info_df["Total Purchase Value"].map("${:,.2f}".format) 
merge_player_info_df["Avg Total Purchase per Person"] = merge_player_info_df["Avg Total Purchase per Person"].map("${:,.2f}".format) 

# Drop the unneeded columns
del merge_player_info_df['Total Count']
del merge_player_info_df['Percentage of Players']

merge_player_info_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create the bins in which Data will be held
bins = [0, 9, 14, 19, 24, 29, 34, 39, 200]

# Create the names for the five bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+" ]


purchase_data_df["Age Range"] = pd.cut(purchase_data_df["Age"], bins, labels=group_names, include_lowest=True)
purchase_data_df

# Create df with just gender information
player_age_df = purchase_data_df[["Age Range","SN"]]
# player_gender_df.head()

# Create a dataframe grouped by gender.
player_age_group = player_age_df.groupby(["Age Range"])
# player_age_group.head()

# Create summation of purchases by gender
player_age_comparison_df = player_age_group.nunique()
# player_age_comparison_df.head()

# Add the total of the column to each entry
player_age_comparison_df["Total"] = player_age_comparison_df["SN"].sum(axis=0)
# player_age_comparison_df.head()


# Clean up the Table column names
# Rename the summary statistic
player_age_comparison_df=player_age_comparison_df.rename(columns = {'SN':'Total Count'})

# Renmove the index name as shown in the sample (just to be exact!!)
player_gender_comparison_df.index.name = None

# Add the Percentage and format it
player_age_comparison_df["Percentage of Players"] = ((player_age_comparison_df["Total Count"] / player_age_comparison_df["Total"])*100)
player_age_comparison_df["Percentage of Players"] = player_age_comparison_df["Percentage of Players"].astype(float).map("{:,.2f}%".format)
player_age_comparison_df.head()

# Drop the Total column
del player_age_comparison_df['Total']

player_age_comparison_df.head(10)
```
